```python
# Filters for different languages
def filters_voices_by_language(voices, language_prefix):
    return [voice for voice in voices if voice.startswith(language_prefix)]

# Filters
english_voices = filters_voices_by_language(VOICES, "en")
french_voices = filters_voices_by_language(VOICES, "fr")
spanish_voices = filters_voices_by_language(VOICES, "es")
portuguese_voices = filters_voices_by_language(VOICES, "pt-BR")

# Results
print("English Voices:", english_voices)
print("French Voices:", french_voices)
print("Spanish Voices:", spanish_voices)
print("Portuguese Voices:", portuguese_voices)
```

```python
topics = ["love", "life", "inspiration", "friendship", "happiness", "success", "wisdom", "motivation"]
quotes = quote(random.choice(topics), limit=1)

for q in quotes:
    TEXT = f"{q['quote']} - {q['author']}"
    print(TEXT)
```

```python
VOICE = random.choice(english_voices)
```

```python
# Output directory
output_dir = "../data/nlp/text-to-speech/"

# Pattern to find existing files
pattern = re.compile(rf"audio-{VOICE}-(\d+)\.mp3")

# Find the highest existing file number
max_num = 0
for filename in os.listdir(output_dir):
    match = pattern.match(filename)
    if match:
        num = int(match.group(1))
        if num > max_num:
            max_num = num

# Increment the number for the next file
next_num = max_num + 1

# Generate the filename dynamically
OUTPUT_FILE = f"{output_dir}audio-{VOICE}-{next_num}.mp3"
print(OUTPUT_FILE)
```

```python
async def amain() -> None:
    communicate = edge_tts.Communicate(TEXT, VOICE)
    await communicate.save(OUTPUT_FILE)

await amain()
```
